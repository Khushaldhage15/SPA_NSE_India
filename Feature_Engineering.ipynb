```python
import pandas as pd
import numpy as np 
import seaborn as sb
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("../DataSets/Stockslive.csv")
df
```

```python
df.drop("Unnamed: 0",axis=1,inplace=True)
df
```

# Daily_return = Close - Open / Open * 100


```python
df["Daily_return"] = df.Close - df.Open / df.Open * 100
df
```

# Price_range = High - Low

```python
df["Price_range"] = df.High - df.Low
df
```

# Volatility = High - Low / Close * 100

```python
df["Volatility"] = df.High - df.Low / df.Close * 100
df
```

# Cumulative_returns = Adj Close.pct_change().cumsum()

```python
df["Cumulative_returns"] = df['Adj Close'].pct_change().cumsum()
df
```

```python
df.Cumulative_returns.fillna(0,inplace=True)
df
```

# Average_price = Open + Close / 2

```python
df["Average_price"] = df.Open + df.Close / 2
df
```

# SMA_50 = Close.rolling(window=50).mean()

```python
df["SMA_50"] = df.Close.rolling(window=50).mean()
df
```

```python
df.SMA_50.fillna(0,inplace=True)
df
```

# SMA_200= Close.rolling(window=200).mean()

```python
df["SMA_200"] = df.Close.rolling(window=200).mean()
df.SMA_200.fillna(0,inplace=True)
df
```

# Avg_price = Open + High + Low + Close / 4


```python
df["Avg_price"] = df.Open + df.High + df.Low + df.Close / 4
df
```

# Volatility_ratio = price_range / Avg_price

```python
df["Volatility_ratio"] = df.Price_range / df.Avg_price
df
```

# Cumulative_value = np.Sum (volume)

```python
df["Cumulative_value"] = np.sum(df.Volume)
df
```

# Price_oscillator = SMA_50 - SMA_200

```python
df["Price_oscillator"] = df.SMA_200- df.SMA_50
df
```

# Volume_weighted_price = close * volume / volume

```python
df['Volume_weighted_price'] = df.Close * df.Volume / df.Volume
df
```
